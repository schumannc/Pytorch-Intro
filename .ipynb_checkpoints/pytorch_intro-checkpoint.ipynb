```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

import numpy as np
np.random.seed(42)

import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='white', rc={"figure.figsize": [10, 10]})
```

# Data

```python
X, y = datasets.make_circles(n_samples=300, noise=0.09, factor=0.2)
X = X.astype(np.float32)
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
```

```python
batch_size = 10

dataset = TensorDataset(torch.from_numpy(X), torch.from_numpy(y))
dataloader = DataLoader(dataset, shuffle=True, batch_size=batch_size)
```

# Network

```python
from torch import nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        D_in = 2
        D_out = 2
        H = 4
        self.fc1 = nn.Linear(D_in, H)
        self.fc2 = nn.Linear(H, D_out)
        
    def forward(self, x):
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        return x
```

```python
model = Net()
print(model)
```

```python
model(torch.from_numpy(X[:10, :]))
```

```python
lr=0.03

# binary cross entropy loss
loss_func = nn.CrossEntropyLoss()
opt = torch.optim.Adam(model.parameters(), lr=lr)
```

```python
epochs = 4

for epoch in range(epochs):
    model.train()
    for xb, yb in dataloader:
        yhat = model(xb)
            
        opt.zero_grad()
        # calculate the loss and perform backprop
        loss = loss_func(yhat, yb)
        loss.backward()
        opt.step()
```

```python
def predict(x, model):
    with torch.no_grad():
        return F.softmax(model(x), dim=1).numpy()
```

```python
preds = predict(torch.from_numpy(X), model)
```

```python
x = np.arange(-2.0, 2.0, 0.1)
y = x

xv, yv = np.meshgrid(x, y)
```

```python
XV = np.column_stack([xv.flatten(), yv.flatten()])
```

```python
decision = predict(torch.from_numpy(XV.astype(np.float32)), model)
```
